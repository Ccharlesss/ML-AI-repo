```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_style('whitegrid')
```

```python
### 1) Create a function that fetches the dataset and transform it in a pd.DataFrame:
from sklearn.datasets import load_breast_cancer
def fetch_dataset() -> pd.DataFrame:
    # Load the dataset
    cancer = load_breast_cancer()
    
    # Create column names by appending 'target' to the feature names
    col_names = list(cancer.feature_names)
    col_names.append('target')
    
    # Create the DataFrame using numpy to concatenate the data and target
    df = pd.DataFrame(np.c_[cancer.data, cancer.target], columns=col_names)
    
    return df

# Fetch the dataset and display the first few rows
df = fetch_dataset()
df.head()
```

```python
### 2) Get some statistics about the pd.DataFrame:
df.describe()
```

```python
### 3) Get some information about columns present in the pd.DataFrame:
df.info()
```

```python
### 4) Get an insight of the relationship of features:
### scatter plot shows how the two features relate to each other (e.g., linear, non-linear, or no clear relationship).
### Diagonal of the pairplot (optional), it displays the distribution of each individual feature.
sns.pairplot(df, hue='target', vars=['mean radius', 'mean texture', 'mean perimeter', 'mean area', 
                                     'mean smoothness', 'mean compactness', 'mean concavity',
                                     'mean concave points', 'mean symmetry', 'mean fractal dimension'])
```

```python
### 5) Display Box plot with nbr of Targets:
sns.countplot(x=df['target'], label = "Count")
```

```python
# Let's check the correlation between the variables 
# Strong correlation between the mean radius and mean perimeter, mean area and mean primeter
plt.figure(figsize=(20,10)) 
sns.heatmap(df.corr(), annot=True) 
```
